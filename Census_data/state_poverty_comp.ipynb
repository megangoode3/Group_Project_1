```python
# import dependencies
from api_keys import census_key
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from census import Census
import hvplot as hv
```

```python
# get poverty data from census API
base_url = "https://api.census.gov/data/2021/acs/acs5/cprofile?get=group(CP03)"
key = "&key=" + census_key
all_states = "&for=state:*"
response = requests.get(base_url + key + all_states).json()
response
```

```python
# put state and poverty data into a dataframe
state_df = pd.DataFrame(response)
state_df.columns = state_df.iloc[0]
state_df = state_df.drop(state_df.index[0])
state_df = state_df.rename(columns={"NAME": "State"})
# add only needed columns to dataframe
state_poverty_df = state_df[["State", "CP03_2021_128E","CP03_2021_009E"]]
state_poverty_df
```

```python
# sort data by state in alphabetical order
state_poverty_df = state_poverty_df.sort_values(by="State")
# rename columns
state_poverty_df = state_poverty_df.rename(columns={"CP03_2021_128E": "Poverty Rate"})
state_poverty_df = state_poverty_df.rename(columns={"CP03_2021_009E": "Unemployment Rate"})
# convert poverty rate to float
state_poverty_df["Poverty Rate"] = state_poverty_df["Poverty Rate"].astype(float)
# convert unemployment rate to float
state_poverty_df["Unemployment Rate"] = state_poverty_df["Unemployment Rate"].astype(float)
state_poverty_df
```

```python
# new dataframe sorted by poverty rate
sorted_state_poverty_df = state_poverty_df.sort_values(by="Poverty Rate", ascending=False)
sorted_state_poverty_df
```

```python
# new dataframe sorted by unemployment rate
sorted_state_unemployment_df = state_poverty_df.sort_values(by="Unemployment Rate", ascending=False)
sorted_state_unemployment_df
```
