```python
# import depenedencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import hvplot.pandas
import requests
from api_keys import weather_api_key
```

```python
# read in csv file
donor_df = pd.read_csv("Weekly_Provisional_Counts_of_Deaths_by_State_and_Select_Causes__2020-2023.csv")
donor_df.head()
```

```python
# Narrow down the data to only the columns we need
deathcounts_df = donor_df[['Jurisdiction of Occurrence', 'MMWR Year', 'MMWR Week','Alzheimer disease (G30)']]
deathcounts_df.count()
```

```python
# rename columns and fill in missing data with 0
deathcounts_df = deathcounts_df.rename(columns={'Jurisdiction of Occurrence': 'State', 'MMWR Year': 'Year', 'MMWR Week': 'Week', 'Alzheimer disease (G30)': 'Deaths'})
deathcounts_df = deathcounts_df.fillna(0)
deathcounts_df.head()
```

```python
# remove the rows with US as the state
deathcounts_df = deathcounts_df[deathcounts_df['State'] != 'United States']

# add latitudes and longitudes to the dataframe
```

```python
# separate the data by year
df_2020 = deathcounts_df[deathcounts_df['Year'] == 2020]
df_2021 = deathcounts_df[deathcounts_df['Year'] == 2021]
df_2022 = deathcounts_df[deathcounts_df['Year'] == 2022]
```

```python
# create a plot tracking the number of deaths over time for each state
plot_2020 = df_2020.hvplot.points(x='Week', y='Deaths', by='State', xlabel='Week', ylabel='Deaths', title='Deaths by Week in 2020', width=2000, height=1000)
plot_2020
```

```python
plot_2021 = df_2021.hvplot.points(x='Week', y='Deaths', by='State', xlabel='Week', ylabel='Deaths', title='Deaths by Week in 2021', width=2000, height=1000)
plot_2021
```

```python
plot_2022 = df_2022.hvplot.points(x='State', y='State', geo=True, width=2000, height=1000)
plot_2022
```
